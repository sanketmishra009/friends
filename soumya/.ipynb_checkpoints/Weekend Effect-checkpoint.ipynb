```python
import pandas as pd
import numpy as np
import yfinance as yf
import datetime
```

```python
Investment = int(input('Enter Investment Amount in $ '))
sd = input('Enter Date of Investment in DD-MM-YYYY format: ').split('-')
start_date = datetime.date(int(sd[2]), int(sd[1]), int(sd[0]))
end_date = datetime.datetime.today()
```

```python
spy_data = yf.download('SPY', start= start_date, end= end_date)

# reseting index for easier access of Date
spy = pd.DataFrame(spy_data)
spy.reset_index(inplace=True)
spy.head()
```

```python
#cleaning the data
spy.drop(['High','Low','Adj Close', 'Volume'],axis=1,inplace=True)
spy.head()
```

```python
# Calculating overnight return % by using ((Open-Prev.CLose)/Close)*100 for SPY
spy['Overnight Return %'] = ((spy['Open'].shift(-1) - spy['Close']).shift(1)/spy['Close'])*100
spy['Overnight Return %'] = spy['Overnight Return %'].fillna(0)
```

```python
#Printing Daily overnight returns and total Profit & Loss(AMZN)
print(str(Investment)+'$ invested in SPY Stock on',start_date) 
spy_investment = Investment
spy_profit = 0
spy_profit_list = []
for i in range(1,spy['Date'].count()):
    spy_pnl =  (spy['Overnight Return %'][i]/100)*Investment
    print('PnL on',spy['Date'][i],'is $',spy_pnl)
    spy_profit = spy_profit + spy_pnl
    spy_profit_list = spy_profit
print(spy)
#print('Current Investment Value: ',(spy_investment+spy_profit))
#print('Total PnL % :',((spy_profit/Investment)*100).round(3))
```

```python
mon_ =  excel_data_df[excel_data_df['Day'] == 'Mon']['Overnight Return %']
tue_ =  excel_data_df[excel_data_df['Day'] == 'Tue']['Overnight Return %']
wed_ =  excel_data_df[excel_data_df['Day'] == 'Wed']['Overnight Return %']
thu_ =  excel_data_df[excel_data_df['Day'] == 'Thu']['Overnight Return %']
fri_ =  excel_data_df[excel_data_df['Day'] == 'Fri']['Overnight Return %']
print(mon_)
```

```python
a = mon_.cumsum().iloc[:-1]
b = mon_[:-1]
dates = excel_data_df['Date']
print(f'{a=}, {b=} , {dates=}')
```

```python
a = mon_.cumsum().iloc[:-1]
b = np.arange(len(mon_)-1)

x = tue_.cumsum().iloc[:-1]
y = np.arange(len(tue_)-1)

c = wed_.cumsum().iloc[:-1]
d = np.arange(len(wed_)-1)

k = thu_.cumsum().iloc[:-1]
l = np.arange(len(thu_)-1)

t = fri_.cumsum().iloc[:-1]
u = np.arange(len(fri_)-1)

fig = plt.figure(figsize=(16, 6))
plt.plot(b, a, label="Monday")
plt.plot(y, x, label="Tuesday")
plt.plot(d, c, label="Wednesday")
plt.plot(l, k, label="Thursday")
plt.plot(u, t, label="Friday")


plt.xlabel('Time')
plt.ylabel('Overnight Return %')
plt.title('Weekday effect')
plt.legend(loc = 2)


plt.show()
```
