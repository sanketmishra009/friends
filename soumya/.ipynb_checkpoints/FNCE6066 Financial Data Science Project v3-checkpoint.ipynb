```python
#Import Modules
import numpy as np
import pandas as pd
import yfinance as yf
from datetime import datetime
from scipy import stats
from scipy.stats import ttest_1samp
```

```python
#Module 1: Data Extraction
"""
Input: ticker,start_date,end_date
Output:Dataframe containining stock data containing the columns:
['Date', 'Open', 'High', 'Low', 'Close', 'Adj Close', 'Volume']
"""

def download_data(ticker,start_date,end_date):
    start_date = '2000-01-01'
    end_date = '2023-05-01'
    data = yf.download(ticker, start_date, end_date)
    data = pd.DataFrame(data)
    data=data.reset_index()
    return data
```

```python
#Module 2: Data Processing
"""
#Creating columns for data manipulation
#Input: stock data from yahoo finance in dataframe  
Columns:
['Date', 'Open', 'High', 'Low', 'Close', 'Adj Close', 'Volume']

#Output:Dataframe containining Weekend effect entries to show weekend effect.

Columns:
["TickerName", "Date", "Open","Close","PrevDate",
"DiffinDays","CloseOpenDiff","WeekendReturn"]
"""

def WeekendReturn(data):
    data["Date"] = pd.to_datetime(data["Date"])
    data = data.drop(["High","Low","Adj Close","Volume"],axis=1)
    data["TickerName"] = ticker

    data["PrevDate"]=data["Date"].shift(1)
    data["DiffinDays"] = data["Date"]-data["PrevDate"]
    data["CloseOpenDiff"]=data["Open"] - data["Close"].shift(1)
    data["WeekendReturn"] = (data["Open"]/ data["Close"].shift(1)) - 1 
#     data["PosWkdRet"] =  data.loc[data["WeekendReturn"]>0,"WeekendReturn"]
#     data["NegWkdRet"] =  data.loc[data["WeekendReturn"]<0,"WeekendReturn"]
    

    #Drops the NaNs caused by shift
    data = data.iloc[1: , :]
    
    #Isolating the weekend effect by choosing only rows with more than 1 day between trading days.
    data["DiffinDays"]=data["DiffinDays"].apply(lambda x:x.days)
    WkdEff_data = data[data["DiffinDays"]>1].reset_index(drop=True)
    
    WkdEff_data = WkdEff_data.reindex(columns=["TickerName", "Date", "Open","Close","PrevDate"
                             ,"DiffinDays","CloseOpenDiff","WeekendReturn"])
    return WkdEff_data
```

```python
#Module 3: Data Description
#Generates a dataframe containing descriptive statistics on a single dataframe for a specified number
#number of years ending with latest date in dataframe.
#Input: Dataframe contaning weekend return data
#output: Dataframe containing returns of each stock for the period specified


def data_desc(df,years,end_date):
    start_date = pd.to_datetime(end_date) - pd.DateOffset(years= years)
#     print(end_date)
#     print(start_date)
    df = pd.DataFrame(df[df["Date"]> start_date])
    df.loc[:, "RetMult"]  = df["WeekendReturn"] + 1 
    df.loc[:,"StockValue"] = df["RetMult"].cumprod(skipna = True)
    
    #T-test
    df1 = pd.DataFrame(columns = df.describe().columns)
    
    t_stat, p_value = ttest_1samp(df["WeekendReturn"], popmean=0)
    df1.loc["T-test","WeekendReturn"] = t_stat
    df1.loc["p_value","WeekendReturn"] = p_value
    df1.loc["PL","WeekendReturn"] = df["StockValue"].iloc[-1]

    df_desc = pd.concat([df.describe(),df1])
    

    
    return(df_desc)
```

```python
#Module 4: Result Data Frame Extraction for Visualization 
from scipy.stats import ttest_1samp

# DataDesc={}
# df = WeekendReturn(ticker)
# df=df.drop(["CloseOpenDiff","Open","Close","DiffinDays"],axis=1)

def data_by_years(df,years,end_date):
    start_date = pd.to_datetime(end_date) - pd.DateOffset(years= years)
#     print(end_date)
#     print(start_date)
    df = pd.DataFrame(df[df["Date"]> start_date])
    df.loc[:, "RetMult"]  = df["WeekendReturn"] + 1 
    df.loc[:,"StockValue"] = df["RetMult"].cumprod(skipna = True)
    df.loc[:,"Long/Short"] = "Long"
    if df.loc[:,"StockValue"].iloc[-1] < 1:
        df.loc[:,"Long/Short"] = "Short"
        df.loc[:, "RetMult"]  = -df["WeekendReturn"] + 1
        df.loc[:,"StockValue"] = df["RetMult"].cumprod(skipna = True)
    
    #T-test
    df1 = pd.DataFrame(columns = df.describe().columns)
    
    t_stat, p_value = ttest_1samp(df["WeekendReturn"], popmean=0)
    df1.loc["T-test","WeekendReturn"] = t_stat
    df1.loc["p_value","WeekendReturn"] = p_value
    df1.loc["PL","WeekendReturn"] = df["StockValue"].iloc[-1]

    df_desc = pd.concat([df.describe(),df1])
    

    
    return(df)
```

```python
#Module 4: Generate DataFrame for Visualization

ticker_list = list(res_df.sort_values(by="Return")["Ticker"].iloc[0:20])
ListofYears = [1]
start_date = '2017-01-01'
end_date = '2023-05-01'

DataCombined_df = pd.DataFrame()
    
for ticker in ticker_list:
    DataCombined={}

    data = download_data(ticker, start_date, end_date)
    
    if data.empty:
        continue  # Skip to the next ticker
        
    df = WeekendReturn(data)
    df = df.drop(["CloseOpenDiff","Open","Close","DiffinDays"],axis=1)
    
    for years in ListofYears:
        DataCombined[f"years {years}"] =  data_by_years(df,years,end_date)
        DataCombined_df = pd.concat([DataCombined_df,DataCombined[f"years {years}"]])

        
DataCombined_df = DataCombined_df.reset_index().drop(["index"],axis=1)
print(DataCombined_df)

from pathlib import Path  
filepath = Path('./FinDataScienceOuputDataforvisualization.csv')  
filepath.parent.mkdir(parents=True, exist_ok=True)  
DataCombined_df.to_csv(filepath)  


```
